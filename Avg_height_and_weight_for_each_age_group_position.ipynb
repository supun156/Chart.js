```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

# Data provided
data = position_sports_profiles

# Current date for age calculation
current_date = datetime.now()

# Data processing with error handling for missing fields
records = []
for position in data:
    for user in position["users"]:
        dob = datetime(user["dateOfBirth"]["year"], user["dateOfBirth"]["month"], user["dateOfBirth"]["date"])
        age = (current_date - dob).days / 365.25
        # Using .get to safely access height and weight, setting defaults if missing
        height = user.get("height", None)
        weight = user.get("weight", None)

        if height is not None and weight is not None:  # Only include entries with both height and weight
            records.append({
                "Position": position["preferedPositionName"],
                "Age": age,
                "Height": height,
                "Weight": weight
            })

# Convert to DataFrame
df = pd.DataFrame(records)

# Function to shorten position names by extracting abbreviations within parentheses
def shorten_position_name(name):
    if "(" in name and ")" in name:
        # Extract text within parentheses
        return name[name.find("(") + 1:name.find(")")]
    return name  # Return original if no parentheses are found

# Apply function to position names in the DataFrame
df['Position'] = df['Position'].apply(shorten_position_name)

# Group by modified Position and calculate mean values
position_stats = df.groupby("Position").mean()

# print(position_stats)

# Plotting
fig, axes = plt.subplots(3, 1, figsize=(10, 15))

# Average Age
axes[0].bar(position_stats.index, position_stats["Age"])
axes[0].set_title("Average Age by Preferred Position")
axes[0].set_xlabel("Position")
axes[0].set_ylabel("Average Age")

# Average Height
axes[1].bar(position_stats.index, position_stats["Height"])
axes[1].set_title("Average Height by Preferred Position")
axes[1].set_xlabel("Position")
axes[1].set_ylabel("Average Height (cm)")

# Average Weight
axes[2].bar(position_stats.index, position_stats["Weight"])
axes[2].set_title("Average Weight by Preferred Position")
axes[2].set_xlabel("Position")
axes[2].set_ylabel("Average Weight (kg)")

plt.tight_layout()
plt.show()
```
